#### 13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means (integer value).

```python
N = 5
men_means = (54.74, 42.35, 67.37, 58.24, 30.25)
men_std= (4, 3, 4, 1, 5)
ind = np.arange(N)

fig = plt.figure()
ax = fig.add_subplot(111)

x = ind
y = men_means

for i,j in zip(x,y):
    ax.annotate(str(j), xy=(i,j))

plt.bar(ind, men_means, width=0.35, yerr=men_std, color='orange')
    
plt.show()
```

#### 14. Write a Python program to create a stacked bar plot with error bars.

```python
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(1,6,1)    

p1 = plt.bar(ind, menMeans, yerr=menStd, color='red')
p2 = plt.bar(ind, womenMeans, bottom=menMeans, yerr=womenStd, color='green')

plt.ylabel('Scores')
plt.xlabel('Groups')
plt.title('Scores by Group and Gender')

plt.show()
```

```python
ind = np.arange(1,9,1)
groups = ['G1','G2','G3','G4','G5','G6','G7','G8']
array = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]]

df = pd.DataFrame(array, columns=groups)
dft = df.transpose()
#print(df)
#print(dft)

dft.plot(kind='barh', sharey=True, stacked=True, grid=True, table=df, legend=False)

plt.figure(figsize=(50, 50), dpi= 80, facecolor='w', edgecolor='k')

plt.show()
```

#### 17. Write a Python program to add textures (black and white) to bars and wedges. 

```python
patterns = ['|', '+', '-', '*', 'o', 'x', '.', '\\', '/']

fig = plt.figure()

ax = fig.add_subplot(111)
for i in range(len(patterns)):
    ax.bar(i, 3, hatch=patterns[i])
    
plt.show()
```
