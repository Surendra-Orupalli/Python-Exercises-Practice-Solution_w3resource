# Matplotlib Pie Chart: Exercises, Practice, Solution

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

#### 1. Write a Python programming to create a pie chart of the popularity of programming Languages. 

```python
df = pd.read_csv("/Users/surendraorupalli/Desktop/Python_W3R/099_Matplotlib_Done/003_Matplotlib Pie Chart_Done/progs.csv")
print(df)

print("-----------------------------------")

plt.pie(x=df['Popularity'], labels=df['Programming languages'], autopct='%2.2f%%')

plt.axis('equal')

plt.title("Popularity of Programming Languages", fontsize=15, fontweight=400)
plt.show()
```

#### 2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.

```python
df = pd.read_csv("/Users/surendraorupalli/Desktop/Python_W3R/099_Matplotlib_Done/003_Matplotlib Pie Chart_Done/progs.csv")

plt.pie(x=df['Popularity'], labels=df['Programming languages'], explode=[0.1,0,0,0,0,0], autopct='%2.2f%%', shadow=True)

plt.axis('equal')

plt.title("Popularity of Programming Languages" + "\n" + "Worldwide, Oct-2017 compared to a year ago", fontsize=12, fontweight=500)
plt.show()
```

#### 3. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.Make multiple wedges of the pie.
